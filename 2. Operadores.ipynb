# Operadores <br>
Los operadores son usados para realizar **operaciones** en variables y valores. <br><br>
Se dividen en:
1. **Básicos**
*   Aritméticos: + , - , * , / , **
*   Asignación: =
*   Comparación: ==, !=, >, >=, <=

2. **Avanzados**
*   Lógicos: and, or, not
*   Identidad: is, is not

## Módulo
Nos permite obtene el **resto** ó **residuo** de una división entera. <br>

Por ejemplo 7 **MOD** 5 devolverá 2, ya que al realizar la división entera de 7 entre 5 se obtiene 1 como cociente y 2 como residuo. <br>

Modulo_v2.png

````python
print(7 % 5)
````

Si tenemos 19 **MOD** 4 devolverá 3, ya que al realizar la división entera de 19 entre 4 se obtiene 4 como cociente y 3 como resto.

````python
print(19 % 4)
````

**Número par:** <br>
Si un número es múltiplo de 2 es par. <br>
Si el **residuo** de **mod 2** es igual a 0, nuestro número será par.

````python
print(2 % 2)
````

**Número impar:** <br>
Si un número es NO es múltiplo de 2 es impar. <br>
Si el **residuo** de **mod 2** es diferente a 0, nuestro número será impar.

````python
print(517 % 2)
````

# 1.2 Operadores de asignaciones

````python
#@title Tabla operadores de asignación
import pandas as pd
pd.DataFrame(columns=['Operador', 'Nombre', 'Ejemplo'],
             index=[1, 2, 3, 4, 5, 6, 7],
             data=[['=', 'igual a', 'mi_variable = 150'],
                   ['+=', 'más igual a', 'x = x + 4'],
                   ['-=', 'menos igual a', 'x = x - 4'],
                   ['*=', 'multipliación igual a', 'x = x * 4'],
                   ['/=', 'división igual a', 'x = x / 4'],
                   ['%=', 'módulo igual a', 'x %= y'],
                   ['**=', 'exponenciación igual a', 'x **= y']])
````

## Operador = <br>
Asigna a la variable (izquierda) un valor (derecha).

x = 1

````python
x = 1
print(x)
````

## Operador +=
El operador += suma y asigna a la variable (izquierda) la adición de la variable con el el valor (derecha).

Esto es:

````python
x = 6
x = x + 4
print(x)
````

es igual a:

````python
x = 6
x += 4
print(x)
````

Tendrá más sentido si revisas las dos opciones en la siguiente celda. ¿Qué opción te gusta más?

````python
# Ejemplo
contador = 1
for n in range(10):
  print(contador)
  contador = contador + 1 # Opción 1
````

````python
# Ejemplo
contador = 1
for n in range(10):
  print(contador)
  contador += 1 # Opción 2
````

Este operador también puede ser usado con cadenas de caracteres (string)

````python
texto = 'Hola'
texto = texto + '_' # 'Hola' + '_'
texto = texto + 'Mundo' # 'Hola_' + 'Mundo'
print(texto)
print(type(texto))
````

````python
texto = 'Hola'
texto += '_'  # 'Hola' + '_'
texto += 'Mundo' # 'Hola_' + 'Mundo'
print(texto)
print(type(texto))
````

# 1.3 Operadores de comparación <br>
* Son usados para comparar dos valores
*   Al usar estos operadores obtendremos un boleano: True, False
*   Si la comparación es *verdadera* regresa **True** y **False** si es falsa.


````python
#@title Tabla operadores de comparación { run: "auto", vertical-output: true }
import pandas as pd
pd.DataFrame(columns=['Operador', 'Nombre', 'Ejemplo'],
             index=[1, 2, 3, 4, 5, 6],
             data=[['==', 'Igual que', 'x == y'],
                   ['!=', 'No es igual que', 'x != y'],
                   ['>', 'Mayor que', 'x > y'],
                   ['>=', 'Mayor o igual que', 'x >= y'],
                   ['<', 'Menor que', 'x < y'],
                   ['<=', 'Menor o igual que', 'x <= y']])
````

## Operador **==**

En el siguiente ejemplo, estamos comparando si el valor 1 es igual que el valor 1. <br>Sabemos 1 es igual a 1, por ello el resultado es True

````python
print(1 == 1)
print(type(1 == 1))
````

**Recuerda que** al utilizar operadores de comparación nos devolveran un valor de tipo **boleano**



Revisa el siguiente ejemplo. Cuál crees que será el resultado: True o False

````python
print(1 == 2)
print(type(1 == 2))
````

Al ser falsa la afirmación, arroja un False.

````python
a, b, c = 2, 3, 3
# 2 == 3
print(a == b)
````

````python
# 3 == 3
print(b == c)
````

````python
# 3 == 2
print(c == a)
````

````python
resultado = c == a
print(resultado)
````

El operador igual que también es usado con cadenas de caracteres:

Ya que las dos palabras no son iguales, arrojara un False

# 1.4 Operadores Lógicos

Nos ayudan a aevaluar múltiples condiciones.

**Palabras clave:** `and`, `or` y `not`

````python
#@title Tabla operadores de lógicos { run: "auto", vertical-output: true }
import pandas as pd
from IPython.display import display, HTML

def pretty_print(df):
    return display(HTML(df.to_html().replace("\\n","<br>") ) )

df = pd.DataFrame(columns=['Operador', 'Descripción', 'Ejemplo'],
             index=[1, 2, 3],
             data=[['and', 'True si ambas evaluaciones son True', 'True and True == a True \\n True and False == False'],
                   ['or', 'True si alguna evaluación es True', 'True or True == False \n True or False == True'],
                   ['not', 'False si alguna evaluación es False', 'True or False == False, \n False or False']])

pretty_print(df)
````
