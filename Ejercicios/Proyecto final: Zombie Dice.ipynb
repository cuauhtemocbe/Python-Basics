## Zombie dice

Eres un zombie con hambre de cerebros. Tu objetivo es obtener 13 cerebros para ganar el jugo. <br>

Para conseguir cerebros deberás tirar los dados que tienen las siguietnes caras: cerebros, huellas y disparos.

El juego contiene 3 tipos de víctimcas (dados) que perseguiras: 
- verdes: humanos fáciles de cazar (más cerebros pocos disparos). Sus caras son: 3 cerebros, 2 huellas y un disparo.
- amarillos: humanos medianamente fáciles de cazar (menos cerebros y más disparos): 2 cerebros, 2 huellas, 2 disparos.
- rojos: humanos muy difíciles de cazar (pocos cerebros y muchos disparos). Sus caras son: 3 disparos, 2 huellas, 1 cerebro.

Tomas 3 dados (de los 13 disponibles) y los tiras. Si obtienes un cerebro, obtienes un punto, pero si obtienes un disparo pierdes un punto de vida. 

Si saca unas huellas, su objetivo habrá escapado y ese dado se podrá volver a lanzar.

**Contenido:**

- 13 dados: 6 verdes, 4 amarillos, y 3 rojos
