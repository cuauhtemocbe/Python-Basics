## Positional Arguments

<br>

Son los argumentos en los que el orden importa.

```python
# Ejemplo
def resta(x, y):
  print(f'x: {x} ; y: {y}')
  print(f'Resultado = {x - y}')
```

```python
resta()
```

```python
resta(1, 2)
```

```python
resta(2, 1)
```

**Cosas importantes para recordar**

- Los argumentos **default** deber铆an ir despu茅s de los argumentos **no-default**
- Argumentos **keyword** deber铆an ir despu茅s de los **positional**
- Los argumentos **default** son opcionales


```python
def resta(y=5, x):
    return x - y
```

```python
def resta(x, y=5):
    return x - y
```

```python
resta(x=5, 2)
```

```python
resta(2, x=5)
```

```python
resta(2, y=5)
```
